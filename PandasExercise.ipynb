```python
#1.Create a pandas DataFrame from the CSV and from Python lists/dicts. 
#I have extended the provided dataset to include 1000 rows
import pandas as pd
df=pd.read_csv("modified_sample.csv")
data=[{ "transaction_id":2001,"store_id":3,"store_name":"Airport","date":"2025-10-05","product_id":507,"product_name":"Scissors","quantity":3,"unit_price":3.50,"total":10.5,"customer_id":2004,"promo_code":"Nan"},{
        "transaction_id": 2002,
        "store_id": 2,
        "store_name": "Downtown",
        "date": "2025-10-06",
        "product_id": 512,
        "product_name": "Notebook",
        "quantity": 5,
        "unit_price": 2.00,
        "total": 10.0,
        "customer_id": 2005,
        "promo_code": "SAVE5"
    },
    {
        "transaction_id": 2003,
        "store_id": 1,
        "store_name": "Mall",
        "date": "2025-10-06",
        "product_id": 520,
        "product_name": "Pen",
        "quantity": 10,
        "unit_price": 1.00,
        "total": 10.0,
        "customer_id": 2006,
        "promo_code": "NaN"
    }]
new_df=pd.DataFrame(data)
df=pd.concat([df,new_df],ignore_index=True)
print(df.tail())
```

```python
#2.Filter and subset rows/columns (e.g., sales for a store, date ranges, high-value transactions).
#Filtering sales for a specific store
storeSales=df.loc[df['store_name']=="Suburb",["transaction_id", "date", "total","promo_code"]]
print("Sales for a Suburb store\n",storeSales)
#filtering transactions based on date ranges
df["date"]=pd.to_datetime(df['date'])
dfDate_filtered=df.loc[(df["date"]>="2025-10-03") & (df["date"]<="2025-10-06"),["transaction_id","total"]]
print("Transactions within a date range:\n",dfDate_filtered)
print(df['total'].max())
highValTransactions=df.loc[df["total"]>=98.8,["transaction_id","store_name"]]
print("High value transactions in the dataset: \n",highValTransactions)
#Combining multiple filters
dfNew=df.loc[(df["store_name"]=="Downtown") & (df["total"]>=4.0),["transaction_id","date","total","customer_id"]]
print("High value sales from Downtown store:\n",dfNew) 
```

```python
#3.Compute descriptive statistics and group summaries (totals by store, average basket). 
#total sales by store
storeTotals = df.groupby('store_name')['total'].sum().reset_index()
storeTotals = storeTotals.sort_values(by='total', ascending=False)
print(storeTotals)
#total quantity sold per product
prodQuant=df.groupby('product_name')['quantity'].sum().reset_index()
prodQuant=prodQuant.sort_values(by='quantity',ascending=False)
print(prodQuant)
```

```python
#4.Perform simple cleaning (handle missing values, fix data types, drop duplicates). 
df['promo_code']=df['promo_code'].fillna('No Promo') #fill missing promo codes with- no promo
print(df['promo_code'])
df['date']=pd.to_datetime(df['date'])#convert date into proper datetime format
print(df['date'])
```

```python
#5.Build a mini ETL pipeline: read CSV → clean & transform → output JSON. 
df.to_json("cleaned_data.json", orient="records", indent=4, date_format="iso")
print("Data Exported to cleaned_data.json Successfully!") 
```

```python
#6.Complete a short hands-on test and an assignment to extend the pipeline.
import io
csv_data_extended = """transaction_id,store_id,store_name,date,product_id,product_name,quantity,unit_price,total,customer_id,promo_code
1001,1,Downtown,2025-10-01,501,Notebook,2,3.5,7,,
1002,1,Downtown,2025-10-01,502,Pen,5,0.8,4,2002,DISC10
1003,2,Suburb,2025-10-02,501,Notebook,1,3.5,3.5,2003,
1004,2,Suburb,2025-10-02,503,Stapler,1,5.25,5.25,,
1005,1,Downtown,2025-10-03,504,Folder,3,1.2,3.6,2004,DISC10
1006,3,Airport,2025-10-03,502,Pen,10,0.8,8,2005,
1007,3,Airport,2025-10-04,505,Marker,2,1.75,3.5,2006,
1008,1,Downtown,2025-10-04,502,Pen,2,0.8,1.6,2002,
1009,2,Suburb,2025-10-05,506,Highlighter,4,0.9,3.6,2007,
1010,2,Suburb,2025-10-05,501,Notebook,1,3.5,3.5,2008,DISC5
# --- New Rows Added ---
1011,4,Online,2025-10-05,507,Mouse,1,15,15,2009,ONLINE20
1012,1,Downtown,2025-10-05,503,Stapler,3,5.25,15.75,2010,
1013,2,Suburb,2025-10-06,505,Marker,1,1.75,1.75,2011,
"""
df = pd.read_csv(io.StringIO(csv_data_extended))
print(f"DataFrame loaded with {len(df)} rows.")
df['date'] = pd.to_datetime(df['date'])
df['customer_id'] = df['customer_id'].fillna(0).astype(int)
highValThresh= 5.00
#Enrichment: add high-value flag
df['is_high_value'] = df['total'] > highValThresh
store_summary_df = df.groupby(['date', 'store_id', 'store_name']).agg(
    **{
        'Total_Revenue': ('total', 'sum'),
        'Transaction_Count': ('transaction_id', 'count'),
        'Avg_Basket_Value': ('total', 'mean'),
        'High_Value_Count': ('is_high_value', 'sum')
    }
).reset_index()
#Aggregation/Metrics: calculate total daily revenue and other metrics
store_summary_df['Total_Revenue'] = store_summary_df['Total_Revenue'].round(2)
store_summary_df['Avg_Basket_Value'] = store_summary_df['Avg_Basket_Value'].round(2)
print("\n--- Daily Store Summary with Extended Data ---")
print(store_summary_df)
output_file = 'cleaned_data.json'
store_summary_df.to_json(
    output_file,
    orient='records',
    date_format='iso',
    indent=4
)
print(f"\nPipeline completed. Summary saved to: {output_file}")
```
